```python
#!pip install pmlb
import pandas as pd
import numpy as np
from pmlb import fetch_data
```

```python
# dataset for this demo
#https://github.com/EpistasisLab/penn-ml-benchmarks/tree/master/datasets/classification/GAMETES_Epistasis_2-Way_20atts_0.4H_EDM-1_1
gametes = fetch_data('GAMETES_Epistasis_2-Way_20atts_0.4H_EDM-1_1')
print(gametes.describe())
```

```python
gametes.head()
```

```python
X=gametes.drop('target',axis=1)

# features for A
A_feat_list = [
                'N10',
                'N11',
                'N12'
                ]

# features for C
C_feat_list = [
                'N13',
                'N14',
                'N15'
                ]
# y

y = gametes['target']
```

```python
# test metaestimator alone
from tpot.builtins import MetaEstimator
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score, train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
metaest = MetaEstimator(estimator=GradientBoostingRegressor(random_state=42), A=A_feat_list, C=C_feat_list)
metaest.fit(X_train, y_train)
metaest.values_list
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train.head()
```

```python
y_pred = metaest.predict(X_train)
```

```python
from tpot.metrics import balanced_accuracy
b_acc = balanced_accuracy(y_true=y_train, y_pred=y_pred)
print(b_acc)
```

```python
print(metaest.score(X_test, y_test))
```

```python
# metaestimator is passed when using it as a stand-alone estimator
```

```python
# test metaestimator with cross_eval_score
cv_scores = cross_val_score(metaest, X, y, cv=3)
print(cv_scores)
```

```python
# test MetaClassifier for binary output
from tpot.builtins import MetaClassifier
metaest2 = MetaClassifier(estimator=GradientBoostingRegressor(random_state=42), A=A_feat_list, C=C_feat_list)

cv_scores = cross_val_score(metaest2, X, y, cv=3, error_score="raise")
print(cv_scores)
```

```python
# score is matched
```

```python
# make a 3-class classification
y_012 = y
y_1_index = np.where(y_012 == 1)

# make 1st 1/3 of class 1 to class 2
size_2 = int(y_1_index[0].size/3)
y_012[y_1_index[0][:size_2]] = 2
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_012, test_size=0.33, random_state=42)
X_train.head()
```

```python
# Test metaestimator with TPOT with config

from tpot import TPOTClassifier


tpot_config = {
    'tpot.builtins.FeatureSetSelector': {
        'subset_list': ['./subset_test.csv'],
        'sel_subset': [0, 1],
        'res_cols': [['N10', 'N11', 'N12', 'N13', 'N14', 'N15']]
    },
   'tpot.builtins.MetaClassifier': {
        
        'estimator': {
           'sklearn.ensemble.GradientBoostingRegressor': {
                'n_estimators': [100],
                'loss': ["ls", "lad", "huber", "quantile"],
                'learning_rate': [1e-3, 1e-2, 1e-1, 0.5, 1.],
                'max_depth': range(1, 11),
                'min_samples_split': range(2, 21),
                'min_samples_leaf': range(1, 21),
                'subsample': np.arange(0.05, 1.01, 0.05),
                'max_features': np.arange(0.05, 1.01, 0.05),
                'alpha': [0.75, 0.8, 0.85, 0.9, 0.95, 0.99]
            },

        },
       'A': [A_feat_list],
       'C': [C_feat_list]
    }
    
}


tpot = TPOTClassifier(generations=5, population_size=25, 
                        verbosity=3, cv=5,
                      config_dict=tpot_config,
                      template="FeatureSetSelector-Classifier",
                      scoring="balanced_accuracy",random_state=42)
tpot.fit(X_train, y_train)
```

```python
list(tpot.evaluated_individuals_.keys())[0]
```

```python
print(tpot.score(X_test, y_test))
```
